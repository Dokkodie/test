```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the surface
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='jet')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some random data
data = np.random.rand(10, 10)

# Create the plot
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='inferno')

# Add the colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Add title and axis labels
ax.set_title('Heatmap')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Add grid lines
ax.grid(False)

# Add tick marks
ax.set_xticks(np.arange(data.shape[1]))
ax.set_yticks(np.arange(data.shape[0]))
ax.set_xticklabels(['Label {}'.format(i) for i in range(data.shape[1])])
ax.set_yticklabels(['Label {}'.format(i) for i in range(data.shape[0])])
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        text = ax.text(j, i, '{:.2f}'.format(data[i, j]),
                       ha='center', va='center', color='w')

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the grid
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create the plot
fig, ax = plt.subplots()
strm = ax.streamplot(X, Y, U, V, color=U, linewidth=2, cmap='viridis')

# Add a colorbar
fig.colorbar(strm.lines)

# Add title and axis labels
ax.set_title('Streamplot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

```python
import plotly.graph_objs as go
import numpy as np

# Create the data
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
z = np.random.rand(N)

# Create the plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers',
                                   marker=dict(size=12, color=z, colorscale='Viridis', opacity=0.8))])

# Add title and axis labels
fig.update_layout(title='3D Scatter Plot', scene=dict(xaxis_title='X',
                                                      yaxis_title='Y',
                                                      zaxis_title='Z'))

fig.show()
```

```python
import plotly.graph_objs as go
import pandas as pd

# Load the data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

# Create the plot
fig = go.Figure(data=go.Scatter(x=df['gdpPercap'],
                                y=df['lifeExp'],
                                mode='markers',
                                marker=dict(size=df['pop']/1e6,
                                            sizemode='area',
                                            sizeref=100000,
                                            color=df['continent'],
                                            colorscale='Portland',
                                            opacity=0.7),
                                text=df['continent']))

# Add title and axis labels
fig.update_layout(title='Bubble Chart', xaxis_title='GDP per Capita',
                  yaxis_title='Life Expectancy')

fig.show()
```

```python
import plotly.graph_objs as go
import numpy as np

# Create the surface
N = 50
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# Create the plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Viridis')])

# Add title and axis labels
fig.update_layout(title='Surface Plot', scene=dict(xaxis_title='X',
                                                   yaxis_title='Y',
                                                   zaxis_title='Z'))

fig.show()
```
